```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df = pd.read_csv("data/author_consistency_report.csv")

df
```

```python
plt.figure(figsize=(10,6))

ax = sns.barplot(
    data=df,
    x="author",
    y="total_commits"
)

plt.title("Número de Commits por Autor")
plt.xlabel("Autor")
plt.ylabel("Total de Commits")

# agregar número de commits en cada barra
for i, row in df.iterrows():
    ax.text(
        i,                       # posición x
        row["total_commits"],    # posición y
        int(row["total_commits"]),
        ha='center',
        va='bottom'
    )

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.barplot(
    data=df,
    x="author",
    y="consistency_score"
)

plt.title("Consistencia de Commits por Autor")
plt.xlabel("Autor")
plt.ylabel("Consistency Score")

plt.ylim(0,1)

plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=df,
    x="total_commits",
    y="consistency_score",
    size="total_commits",
    legend=False
)

plt.title("Relación entre Experiencia y Consistencia")
plt.xlabel("Total Commits")
plt.ylabel("Consistency Score")

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.barplot(
    data=df,
    x="author",
    y="avg_lines_changed"
)

plt.title("Tamaño Promedio de Commits")
plt.xlabel("Autor")
plt.ylabel("Líneas Modificadas")

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.barplot(
    data=df,
    x="author",
    y="size_variation"
)

plt.title("Variabilidad del Tamaño de Commit")
plt.xlabel("Autor")
plt.ylabel("Desviación Estándar")

plt.show()
```

```python
ranking = df.sort_values(
    "consistency_score",
    ascending=False
)

ranking[["author", "consistency_score"]]
```

```python
plt.figure(figsize=(10,6))

sns.barplot(
    data=ranking,
    x="consistency_score",
    y="author"
)

plt.title("Ranking de Consistencia de Desarrolladores")

plt.show()
```
