```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set_style("whitegrid")
```

```python
df = pd.read_csv("data/author_consistency_report.csv")

df
```

```python
# Configurar estilo
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

fig, ax = plt.subplots(figsize=(12, 6))

# Gráfico de barras horizontal
bars = ax.barh(df["author"], df["total_commits"])

# Personalizar colores
colors = plt.cm.viridis(np.linspace(0, 1, len(df)))
for bar, color in zip(bars, colors):
    bar.set_color(color)

# Títulos y etiquetas
ax.set_title("Número de Commits por Autor", fontsize=16, fontweight='bold', pad=20)
ax.set_xlabel("Total de Commits", fontsize=12)
ax.set_ylabel("Autor", fontsize=12)

# Agregar valores en las barras
for i, (author, commits) in enumerate(zip(df["author"], df["total_commits"])):
    ax.text(commits + 0.1, i, f'{int(commits)}', 
            va='center', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.barplot(
    data=df,
    x="author",
    y="consistency_score"
)

plt.title("Consistencia de Commits por Autor")
plt.xlabel("Autor")
plt.ylabel("Consistency Score")

plt.ylim(0,1)

plt.show()
```

```python
df = pd.read_csv("data/author_consistency_report.csv")

df = df.sort_values("total_commits", ascending=False)

plt.figure(figsize=(10,6))

ax = sns.barplot(
    data=df,
    x="author",
    y="total_commits"
)

plt.title("Número de Commits por Autor")
plt.xlabel("Autor")
plt.ylabel("Total de Commits")

for container in ax.containers:
    ax.bar_label(container)

plt.show()
```

```python
plt.figure(figsize=(10,6))

ax = sns.barplot(
    data=df,
    x="author",
    y="consistency_score"
)

plt.title("Consistencia de Commits por Autor")
plt.xlabel("Autor")
plt.ylabel("Consistency Score")

plt.ylim(0,1)

for container in ax.containers:
    ax.bar_label(container)

plt.show()
```
